# Load MNIST

```python
from keras.datasets import mnist
import matplotlib.pyplot as plt

(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# Plot some images
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Number: {y_train[i]}")
    plt.axis("off")
plt.show()
```

```python
# Check data shape
print("Train:", "X =", X_train.shape, "; y =", y_train.shape)
print("Test:", "X =", X_test.shape, "; y =", y_test.shape)
```

```python
# Reshape feature matrix
X_train = X_train.reshape(-1, 28 * 28)
X_test = X_test.reshape(-1, 28 * 28)
```

```python
# Use a small amount of data for fast training
X_train = X_train[:12000]
y_train = y_train[:12000]
X_test = X_test[:2000]
y_test = y_test[:2000]
```

# Compare function

```python
# Helper function to evaluate model's performance
from sklearn.metrics import classification_report
def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print(classification_report(y_true=y_test, y_pred=y_pred))
```

# Load Softmax Regression model from scikit-learn

```python
from sklearn.linear_model import LogisticRegression

model_1 = LogisticRegression(random_state=42, C=10e5)
```

```python
model_1.fit(X_train, y_train)
```

```python
evaluate(model_1, X_test, y_test)
```

# Load Softmax Regression Model trained with Genetic Algorithm

```python
# Load model module
from genetic_classifier import Evolution
```

```python
# Model fit
evo = Evolution(X_train, y_train)
model_2 = evo.start(fitness_threshold=0.86)
```

```python
evaluate(model_2, X_test, y_test)
```
